Q1. Write a Python program to print numbers from 1 to 10 using a while loop.

```python
num = 1
while num<=10:
  print(num)
  num+=1
```

Q2. Write a Python program to calculate the sum of numbers from 1 to 100 using a while loop.

```python
num = 1
sum = 0
while num<=100:
  sum +=num
  num+=1
print("Sum of numbers 1 to 100: ",sum)
```

Q3. Write a Python program to print even numbers from 2 to 20 using a while loop.

```python
num = 2
while num<=20:
  if num%2 == 0:
    print(num)
  num+=1
```

Q4. Write a Python program to find the factorial of a number entered by the user using a while loop.

```python
num = int(input("Enter a number: "))
fac = 1
while num >= 1:
  fac *= num
  num -=1
print("Factiorial is: ",fac)
```

Q5. Write a Python program to print the Fibonacci series up to 50 using a while loop.

```python
num1 = 0
num2 = 1
num3 = 0
print(num1)
print(num2)
while  num3<=50:
  num3 = num1+num2
  num1 = num2
  num2 = num3
  print(num3)
```

Q6 .Write a Python program to find the square of numbers from 1 to 10 using a while loop.

```python
num = 1
while num <= 10:
  print(num ** 2)
  num+=1
```

Q8 .Write a Python program to print numbers from 10 to 1 in reverse order using a while loop.

```python
num = 10
while num >=1:
  print(num)
  num -=1
```

Q9. Write a Python program to count the number of digits in a positive integer entered by the user using a while loop.

```python
num = int(input("Enter a number: "))
cnt = 0

while num != 0:
    num //= 10
    cnt += 1

print("Number of digits: ",cnt)
```

Q10. Write a Python program to find the sum of digits of a positive integer entered by the user using a while loop.

```python
num = int(input("Enter a number: "))
sum = 0

while num != 0:
    sum += num % 10
    num //= 10

print("Sum of digits: ",sum)
```

Q11. Write a Python program to check if a number entered by the user is a palindrome using a while loop.

```python
num = int(input("Enter number: "))
temp = num
rev=0
while num > 0:
    dig=num%10
    rev=rev*10+dig
    num=num//10
if(temp==rev):
    print("The number is palindrome!")
else:
    print("Not Palindrome!")
```

Q13. Write a Python program to reverse a number entered by the user using a while loop.

```python
num = int(input("Enter number: "))
rev=0
while num > 0:
    dig=num%10
    rev=rev*10+dig
    num=num//10
print("Reverse number: ",rev)
```

Q14. Write a Python program to print the multiplication table of a number entered by the user using a while loop.


```python
num = int(input("Enter number: "))
i = 1
while i<=10:
  print(i," * ",num," = ",i*num)
  i +=1
```

Q15. Write a Python program to calculate the power of a number entered by the user using a while loop.

```python
base = int(input("Enter number: "))
exponent = int(input("Enter power: "))
result = 1

while exponent != 0:
     result *= base
     exponent-=1

print("Answer = ",result)
```
