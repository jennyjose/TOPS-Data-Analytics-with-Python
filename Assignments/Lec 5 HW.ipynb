Question 8: Given two dictionaries dict1 = {'a': 1, 'b': 2} and dict2 = {'c': 3, 'd': 4}, how can you perform concatenation on them? Write a Python program to achieve this.

```python
dict1 = {'a': 1, 'b': 2}
dict2 = {'c': 3, 'd': 4}

print("Dictionary1: ",dict1)
print("Dictionary2: ",dict2)

#dict3 = {**dict1,**dict2}

dict3 = dict1 |dict2
print("Concatenated dictionary: ",dict3)
```
