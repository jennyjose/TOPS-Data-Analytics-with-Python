Question 10: Write a Python program that checks if a given string is an English word, a number, or a combination of both. Print the corresponding message.

```python
str3 = input("Enter string: ")
len_alp = 0
len_dig =0
for i in str3:
  if i.isalpha() == True:
    len_alp +=1
  elif i.isdigit() == True:
    len_dig += 1

if len(str3) == len_alp:
  print("String contains English word")
elif len(str3) == len_dig:
  print("String contains numbers")
elif len(str3) == (len_alp+len_dig):
  print("String is combination of both")
```
