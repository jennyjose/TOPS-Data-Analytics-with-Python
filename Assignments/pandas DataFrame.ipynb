Question 5: Summary Statistics
Given the DataFrame df from Question 1, print the summary statistics (mean, min, max, etc.) for numeric columns.



```python
print("Mean of Age: ",df['Age'].mean())
print("Minimum of Age: ",df['Age'].min())
print("Maximum of Age: ",df['Age'].max())
print("Median of Age: ",df['Age'].median())
```

Question 7: Filtering Data
Given the DataFrame df from Question 1, filter and print the rows where Age is greater than 30.

Question 8: Sorting Data
