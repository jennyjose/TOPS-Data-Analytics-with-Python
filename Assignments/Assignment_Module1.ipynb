Q6. Write a Python program to find the first appearance of the substring 'not' and 'poor' from a given string, if
'not' follows the 'poor', replace the whole 'not'...'poor' substring with 'good'.
Return the resulting string

```python
def not_poor(str1):
    snot = str1.find('not')
    spoor = str1.find('poor')

    if spoor > snot and snot > 0 and spoor > 0:
        str1 = str1.replace(str1[snot:(spoor+4)], 'good')
        return str1
    else:
        return str1

print(not_poor('The lyrics is not that poor!'))
print(not_poor('The lyrics is poor!'))
```

Q8. Write a Python program to check whether a list contains a sublist.

```python
def is_Sublist(l, s):
    sub_set = False

    if s == []:
        sub_set = True
    elif s == l:
        sub_set = True
    elif len(s) > len(l):
        sub_set = False
    else:
        for i in range(len(l)):
            if l[i] == s[0]:
                n = 1
                while (n < len(s)) and (l[i + n] == s[n]):
                    n += 1

                if n == len(s):
                    sub_set = True

    return sub_set

list1 = [2, 4, 3, 5, 7]
list2 = [4, 3]
list3 = [3, 7]

print("List1: ",list1)
print("List2: ",list2)
print("List3: ",list3)

print("List2 is a sublit of List1: ",is_Sublist(list1, list2))

print("List3 is a sublist of List1: ",is_Sublist(list1, list3))
```

Q13. Write a Python program to sort a dictionary (ascending /descending) by value.

```python
import operator

d = {1: 2, 3: 4, 4: 3, 2: 1, 0: 0}

print('Original dictionary : ',d)

dict_sort = dict(sorted(d.items(),key=operator.itemgetter(1)))
print('Dictionary in ascending order by value : ',dict_sort)

dict_sort_r = dict( sorted(d.items(),key=operator.itemgetter(1),reverse=True))
print('Dictionary in descending order by value : ',dict_sort_r)
```

Q16. Counting the frequencies in a list using a dictionary in Python.
Input : [1, 1, 1, 5, 5, 3, 1, 3, 3, 1,4, 4, 4, 2, 2, 2, 2]
Expected output : 1 : 5 , 2 : 4 , 3 : 3 , 4 : 3 , 5 : 2

```python
list1 = [1, 1, 1, 5, 5, 3, 1, 3, 3, 1,4, 4, 4, 2, 2, 2, 2]
freq = {}
for item in list1:
  if (item in freq):
    freq[item] += 1
  else:
    freq[item] = 1

dict_sort = dict(sorted(freq.items()))
print(dict_sort)
```

Q17. Write a python program using function to find the sum of odd series and even series
Odd series: 12/ 1! + 32/ 3! + 52/ 5!+……n
Even series: 22/ 2! + 42/ 4! + 62/ 6!+……n

```python
def factorial(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * factorial(n - 1)
# num = int(input("Enter a number = "))
# print("Factorial of ",num," is: ",factorial(num))

def sum_odd(n):
  i = 1
  sum = 0
  k = 0
  s = ""
  while i<=n:
    sum += (12+k)/factorial(i)

    s += " "+str(12+k)+"/"+str(i)+"! +"
    i += 2
    k += 20
  print(s)
  print("Sum: ",sum)

def sum_even(n):
  i = 2
  sum = 0
  k = 0
  s = ""
  while i<=n:
    sum += (12+k)/factorial(i)

    s += " "+str(12+k)+"/"+str(i)+"! +"
    i += 2
    k += 20
  print(s)
  print("Sum: ",sum)
```
