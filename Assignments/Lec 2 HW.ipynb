#Q1- Write a Python program to take two numbers as input from the user and display their sum.


```python
num1 = int(input("Enter num1 = "))
num2 = int(input("Enter num2 = "))
add = num1+num2
print("Addition of ",num1," and ",num2," = ",add)
```

#Q2 - Calculate and print the midpoint of two numbers entered by the user.

```python
num1 = int(input("Enter num1 = "))
num2 = int(input("Enter num2 = "))
mid = (num1+num2)/2
print("Midpoint of ",num1," and ",num2," = ",mid)
```

#Q3 - Write a Python program to calculate the area of a triangle using the formula: area = 0.5 * base * height.

```python
base = int(input("Enter base = "))
height = int(input("Enter height = "))
area = 0.5*base*height
print("Area of triangel with ",base," and ",height," = ",area)
```

#Q4 - Take three numbers as input from the user and display their product.

```python
num1 = int(input("Enter num1 = "))
num2 = int(input("Enter num2 = "))
num3 = int(input("Enter num3 = "))
product = num1*num2*num3
print("Product of ",num1," , ",num2," and " ,num3, " = ",product)
```

#Q5 - Take a number as input and display its square and cube.

```python
num1 = int(input("Enter num1 = "))
square = num1**2
cube = num1**3
print("Square of ",num1," is = ",square)
print("Cube of ",num1," is = ",cube)
```

#Q6 - Take two numbers as input and display their quotient and remainder.

```python
num1 = int(input("Enter num1 = "))
num2 = int(input("Enter num2 = "))
quotient = num1/num2
remainder = num1%num2
print("Quotient of ",num1," and ",num2, " is = ",quotient)
print("Remainder of ",num1," and ",num2, " is = ",remainder)
```

#Q7 - Take three numbers as input and display their average.

```python
num1 = int(input("Enter num1 = "))
num2 = int(input("Enter num2 = "))
num3 = int(input("Enter num3 = "))
average = (num1+num2+num3)/3
print("Average of ",num1," , ",num2," and " ,num3, " = ",average)
```

#Q8 - Write a Python program to calculate the area of a rectangle using the formula: area = length * width.

```python
length = int(input("Enter length = "))
width = int(input("Enter width = "))
area = length*width
print("Area of rectangle with ",length," and ",width," = ",area)
```

#Q9 - Take five numbers as input representing marks obtained and total marks. Display the percentage.

```python
print("Enter marks for 5 subjects out of 100:")
subject1 = int(input("Enter subject1 = "))
subject2 = int(input("Enter subject2 = "))
subject3 = int(input("Enter subject3 = "))
subject4 = int(input("Enter subject4 = "))
subject5 = int(input("Enter subject5 = "))
total_marks = subject1+subject2+subject3+subject4+subject5
percentage = (total_marks/500)*100
print("percentage is ",percentage)
```
