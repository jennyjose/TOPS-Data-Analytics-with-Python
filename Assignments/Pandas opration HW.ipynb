Question 10: Handling Missing Values
Given the DataFrame df_missing from Question 9, replace missing values in the 'Age' column with the median age of the other records and replace missing values in the 'Salary' column with the median salary.

```python
med_age = df_missing['Age'].median()
med_sal = df_missing['Salary'].median()

df_missing['Age'] =  df_missing['Age'].fillna(med_age)
df_missing['Salary'] = df_missing['Salary'].fillna(med_sal)
df_missing
```
